# Machine Learning Model for School Management

```python
import sklearn
import pickle
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
```

## Loading data

```python
df_raw = pd.read_csv('dataset.csv')
df_raw.info()
```

```python
df_raw.sample(5)
```

## Data preprocessing

```python
# We split into X and y
X = df_raw.drop(columns=['admitted'])
y = df_raw['admitted']

# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)

# Creating the standardizer
scaler = StandardScaler()

# Doing the fit and transform only on the train data
X_train_scaled = scaler.fit_transform(X_train)

# Doing the transform on the test data
X_test_scaled = scaler.transform(X_test)
```

## Version 1 - Logistic Regression

Logistic Regression is a machine learning algorithm used to predict the probability of a categorical dependent variable. Although the name suggests regression, it is more commonly used in binary classification problems, where the goal is to predict one of two possible classes (such as "yes" or "no", "1" or "0", "positive" or "negative"). The algorithm models the probability that a given input belongs to a specific category.

The operation of Logistic Regression is based on the concept of a logistic function, also known as a sigmoid function. This function has an "S"-shaped curve, which can map any real number to a value between 0 and 1, making it suitable for estimating probabilities. The model calculates the probability that the input belongs to a specific class, and if this probability is greater than a defined threshold (usually 0.5), the input is classified into that class. The logistic function transforms the linear output (a linear combination of independent variables weighted by coefficients, plus an intercept term) into a probability.

In terms of training, Logistic Regression involves adjusting the coefficients of the independent variables. This is done by maximizing the likelihood function, which seeks to find the coefficients that make the predicted probabilities closest to the observed (real) results. Essentially, the algorithm attempts to draw a decision line that best separates the classes. To evaluate the model's performance, metrics such as accuracy, confusion matrix, AUC-ROC (Area Under the Curve - Receiver Operating Characteristic), and others are used.

Link: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
# Creating the first model
model_v1 = LogisticRegression()


# Training the model
model_v1.fit(X_train_scaled, y_train)
```

```python
# Getting the predictions
y_pred_v1 = model_v1.predict(X_test_scaled)

# Getting the predicted probabilities
y_pred_proba_v1 = model_v1.predict_proba(X_test_scaled)[:, 1]

print(y_pred_v1)
print(y_pred_proba_v1)
```

```python
# Evaluating the accuracy
acc_v1 = accuracy_score(y_test, y_pred_v1)
print(f"\nAccuracy: {acc_v1}")

# Evaluating the AUC
auc_v1 = roc_auc_score(y_test, y_pred_proba_v1)
print(f"\nAUC: {auc_v1}")

# Generating the classifier report
class_report_v1 = classification_report(y_test, y_pred_v1)
print("\nClassifier report:\n", class_report_v1)
```

## Version 2 - Random Forest

The RandomForest algorithm, part of the ensemble learning family, is one of the most popular and versatile machine learning algorithms, used for both classification and regression tasks. Its main characteristic is the construction of an ensemble of decision trees during training, and the model's output is determined by averaging the predictions of all trees in the case of regression, or by majority vote in the case of classification. The name "RandomForest" derives from the way the algorithm creates a "forest" of random decision trees during training.

Each tree in the random forest is built from a random sample of the training data (with replacement), known as bootstrapping, which gives the model the property of being an "ensemble bagging". Furthermore, in each split of a tree, a random subset of features is selected for consideration, increasing the diversity among the trees and making the model more robust against overfitting. This randomness helps improve the model's accuracy and reduce variance, providing an effective solution for handling large and complex datasets. Individual trees tend to learn and adapt excessively to training data (overfitting), but when combined into a random forest, the errors of one tree are compensated for by the successes of the others, resulting in improved overall performance.

RandomForest is widely appreciated for its ease of use, as it requires little hyperparameter configuration and generally produces good results with default settings. It is also less prone to overfitting than a single decision tree. Furthermore, it offers good interpretability through feature importance, which measures how much each feature contributes to the model's prediction accuracy. However, despite its many advantages, it can be computationally intensive and less efficient in terms of time and memory, especially with a very large number of trees or a very large dataset.

Link: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
model_v2 = RandomForestClassifier()
model_v2.fit(X_train_scaled, y_train)
```

```python
y_pred_v2 = model_v2.predict(X_test_scaled)
y_pred_proba_v2 = model_v2.predict_proba(X_test_scaled)[:, 1]

print(y_pred_v2)
print(y_pred_proba_v2)
```

```python
# Evaluating the accuracy
acc_v2 = accuracy_score(y_test, y_pred_v2)
print(f"\nAccuracy: {acc_v2}")

# Evaluating the AUC
auc_v2 = roc_auc_score(y_test, y_pred_proba_v2)
print(f"\nAUC: {auc_v2}")

# Generating the classifier report
class_report_v2 = classification_report(y_test, y_pred_v2)
print("\nClassifier report:\n", class_report_v2)
```

## Version 3 - Support Vector Machine (SVM)

Support Vector Machines (SVM) is a powerful and versatile machine learning algorithm, primarily used for classification tasks, but also applicable in regression. In its simplest and most commonly used form, SVM is employed for binary classification.

The main goal of SVM is to find a hyperplane in an N-dimensional space (N being the number of features) that distinctly classifies data points into two categories. This hyperplane is chosen to be as far away as possible from the nearest data points of each class, known as support vectors. This approach helps maximize the margin between classes, contributing to a more robust model with better generalization capabilities.

One of the main strengths of SVM is its ability to work efficiently in high-dimensional spaces with clear separation margins between classes. However, when the data is not linearly separable, SVM uses a technique called "kernel trick." This technique transforms the data into a higher-dimensional space where linear separation is possible. The most common kernels are the polynomial, the radial basis function (RBF), and the sigmoid. This approach makes SVM particularly effective in complex cases where the relationship between features and classes is not immediately apparent.

Despite its advantages, SVM can be challenging in terms of choosing and adjusting the correct parameters, such as the kernel type and kernel parameters, which have a significant impact on model performance. Furthermore, SVM tends to be less memory-efficient and computationally more intensive, especially with large datasets. It can also be less effective in situations with a high number of features relative to the number of samples or in cases of very unbalanced classes. Despite these challenges, its effectiveness in separating complex classes makes it a popular choice for classification problems in various fields.

Link: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

```python
model_v3 = SVC(verbose=True, probability=True)

model_v3.fit(X_train_scaled, y_train)

y_pred_v3 = model_v3.predict(X_test_scaled)
y_pred_v3
```

```python
y_pred_proba_v3 = model_v3.predict_proba(X_test_scaled)[:, 1]
y_pred_proba_v3
```

```python
# Evaluating the accuracy
acc_v3 = accuracy_score(y_test, y_pred_v3)
print(f"\nAccuracy: {acc_v3}")

# Evaluating the AUC
auc_v3 = roc_auc_score(y_test, y_pred_proba_v3)
print(f"\nAUC: {auc_v3}")

# Generating the classifier report
class_report_v3 = classification_report(y_test, y_pred_v3)
print("\nClassifier report:\n", class_report_v3)
```

### Evaluation Metrics

Let's choose the model with the highest AUC. Here are some reasons to opt for a model with a higher AUC:

**Comprehensive Performance Measure**: The AUC-ROC is a measure that considers all possible classification thresholds, providing an overview of the model's performance at different levels of sensitivity and specificity. A model with a higher AUC generally indicates that it has a better ability to distinguish between positive and negative classes at various thresholds.

**Robustness to Class Imbalance**: In many real-world problems, the number of observations in one class is much larger than the other, which can lead to a biased model. The AUC-ROC is less affected by class imbalance than other metrics, such as precision, because it assesses the model's performance at all classification thresholds, not just the standard cutoff decision.

**Balance Between Sensitivity and Specificity**: Choosing a model with the highest AUC implies that the model not only correctly identifies true positives (sensitivity) but also correctly excludes true negatives (specificity). This is crucial in many practical applications, such as medical diagnosis, where both false negatives and false positives can have serious consequences.

**Model Comparison**: AUC offers a single metric that can be used to compare the performance of different models in a classification problem. This simplifies the choice between models, as a higher AUC generally indicates a model that performs better at correctly classifying observations.

**Cutoff Decision Independence**: Unlike metrics that depend on a specific cutoff point (such as precision, recall, and F1-Score), AUC considers the model's performance at all possible cutoff points. This means that AUC reflects the model's quality in terms of its ability to rank predictions correctly, regardless of how the decision threshold is subsequently chosen.

**Probability Prediction**: The ROC curve and, consequently, the AUC are particularly useful when we are interested in predicting class probabilities rather than binary classifications. A model that produces calibrated and discriminative probabilities between classes will have a higher AUC.
